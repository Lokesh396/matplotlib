```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn')
```

```python
x=[5,7,6,8,4,3,4,6,3,7,5,8,5,6,4,6,3,4,2,4]
y=[4,5,3,5,7,4,7,4,5,8,4,7,4,7,4,5,7,3,4,2]
colors=[5,7,6,8,4,3,4,6,3,7,5,8,5,6,4,6,3,4,2,4]
sizes=[200,300,432,232,323,256,278,253,239,290,390,400,320,230,210,289,345,276,300,450]
```

```python
plt.scatter(x,y,s=sizes,c=colors,cmap='Greens',edgecolor='black',linewidth=1,alpha=0.75)
plt.title('Scatter Plots')
plt.xlabel('participation')
plt.ylabel('Satisfaction')
cbar=plt.colorbar()
cbar.set_label('Satisfaction_index')
plt.savefig('scatter1.png')
plt.show()
```

```python
df=pd.read_csv('scatterdata.csv')
df.head(1)
```

```python
view_count=df['view_count']
likes=df['likes']
ratio=df['ratio']
```

```python
plt.scatter(view_count,likes,c=ratio,cmap='summer',edgecolor='black')
cbar=plt.colorbar()
cbar.set_label('like to dislike Ratio')
plt.xscale('log')
plt.yscale('log')
plt.title('trending youtube videos')
plt.xlabel('View_count')
plt.ylabel('Total likes')
plt.savefig('scatter2.jpg')
plt.show()
```
