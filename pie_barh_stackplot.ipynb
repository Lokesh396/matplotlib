```python
df2=pd.read_csv('langdata.csv')
df2.drop(columns='Unnamed: 0',inplace=True)
```

```python
df2.head(2
        )
```

```python
from matplotlib import pyplot as plot
from collections import Counter
```

```python
idx=df2['Respondent']
langs=df2['LanguageWorkedWith']
```

```python
lang_counter=Counter()
```

```python
for response in langs:
        lang_counter.update(str(response).split(';'))
```

```python
lang_counter.most_common(15)
```

```python
languages=[]
popularity=[]
for item in lang_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
languages,popularity
```

```python
plot.xlabel('popularity')
plot.title('Language popularity')
plot.barh(languages,popularity,color='m')
```

```python

slices=[30,50,20]
labels=['thirty','fifty','twenty']
colors=['r','m','g']
plot.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},colors=colors)
plot.title('PIE Chart')
plot.tight_layout()
```

```python
slices=[59219,
  55466,
  47544,
  36443,
  35917]
languages_pie=['JavaScript',
  'HTML/CSS',
  'SQL',
  'Python',
  'Java']
explode=[0,0,0,0.12,0]
```

```python
plot.title('Language Popularity')
plot.pie(slices,labels=languages_pie,explode=explode,shadow=True,
         wedgeprops={'edgecolor':'black'},autopct='%1.1f%%',
         startangle=100)
plot.savefig('piechart.png')
```

```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
minutes=[1,2,3,4,5,6,7,8,9]
player1=[1,2,3,3,4,4,4,4,5]
player2=[1,2,2,2,3,3,3,3,4]
player3=[1,2,2,2,4,4,4,4,5]
labels=['player1','player2','player3']
plt.title('Stackplot')
plt.stackplot(minutes,player1,player2,player3,labels=labels)
plt.legend(loc='upper left')
plt.show()
```

```python
dev1=[8,6,5,3,3,2,1,0,0]
dev2=[0,1,1,3,2,4,5,4,4]
dev3=[0,1,2,2,3,2,2,4,4]
```

```python
plt.stackplot(minutes,dev1,dev2,dev3,labels=['dev1','dev2','dev3'])
plt.legend(loc=(0.05,0.05))
plt.savefig('slackplot.png')
plt.show()
```
