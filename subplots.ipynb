```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')
```

```python
data=pd.read_csv('subplots.csv')
data.head(1)
```

```python
ages=data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']
```

```python
fig,ax=plt.subplots()
ax.plot(ages,dev_salaries,label='All dev')
ax.plot(ages,py_salaries,label='pydev')
ax.plot(ages,js_salaries,linestyle='--',color='k',label='js')
ax.set_title('Dev Salries Median(USD)')
ax.set_xlabel('Ages')
ax.set_ylabel('Salry USD')
ax.legend()
plt.show()
```
