```python
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
import pandas as pd
```

```python
df=pd.read_csv('linedata.csv')
df.head(2)
```

```python
ages=df['Age']
dev_salaries=df['All_Devs']
py_dev=df['Python']
js_dev=df['JavaScript']
```

```python
plt.title('Median Dev salaries(USD)')
plt.plot(ages,dev_salaries,linestyle='--',color='m',label='Dev_Salaries')
plt.plot(ages,py_dev,color='r',label='PyDev_Salaries')
plt.fill_between(ages,py_dev,dev_salaries,where=(py_dev>dev_salaries),
                label='Above_avg',interpolate=True,alpha=0.2)
plt.fill_between(ages,py_dev,dev_salaries,where=(py_dev<=dev_salaries),
                 label='Below_avg',interpolate=True,alpha=0.2)
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.legend()
plt.savefig('lineplot.png')
plt.show()
```
