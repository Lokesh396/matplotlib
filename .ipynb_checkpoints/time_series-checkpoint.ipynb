```python
import pandas as pd
from datetime import datetime,timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
plt.style.use('seaborn')
```

```python
dates=[
    datetime(2019,5,24),
    datetime(2019,5,25),
    datetime(2019,5,26),
    datetime(2019,5,27),
    datetime(2019,5,28),
    datetime(2019,5,29),
    datetime(2019,5,30),
]

y=[0,1,3,4,6,5,7]
```

```python
plt.plot_date(dates,y,linestyle='solid')

#gcf=git current figure auto formate on xaxis
plt.gcf().autofmt_xdate()
date_fomat=mpl_dates.DateFormatter('%b,%d %Y')
plt.gca().xaxis.set_major_formatter(date_fomat)
plt.savefig('timeseries1.jpg')
plt.show()
```

```python
data=pd.read_csv('time_series_data.csv')
price_data=data['Date']
price_close=data['Close']
```

```python
plt.plot_date(price_data,price_close,linestyle='solid')
plt.gcf().autofmt_xdate()
plt.title('Bitcoin Prices')
plt.xlabel('Dates')
plt.ylabel('Closed Price')
plt.savefig('timeseries2.jpg')
plt.show()
```
